```python
object_info_path = 'data/ADEChallengeData2016/objectInfo150.txt'
object_info = list(csv.reader(open(object_info_path, 'r'), delimiter='\t'))

object_info[:10] # note that '0' represents untagged objects
object_names = [o[4].split(',')[0] for o in object_info]
```

```python
training_image_paths = sorted(glob.glob('data/ADEChallengeData2016/images/training/ADE_train_*.jpg'))
training_annotation_paths = sorted(glob.glob('data/ADEChallengeData2016/annotations/training/ADE_train_*.png'))
```

```python
# just get an idea of the layout of the data
print("    | Image                 | Annotation         |")
print("   #| shape           type  | shape        type  |")
print("----|-----------------------|--------------------|")
for i in range(10):
    img = misc.imread(training_image_paths[i])
    ann = misc.imread(training_annotation_paths[i])
    
    print("{:4}| {}  {}  | {}  {}  |".format(i, img.shape, img.dtype, ann.shape, ann.dtype))
```
